```python
titanic = sns.load_dataset("titanic")
```

```python

sns.kdeplot(data=titanic , x="age" , hue = "survived")

# Between the age 15-19 probablity of surviving is more than killing
```

<h1>Standard Normal Variate = mean-0 std-0
</h1>

```python
sns.kdeplot(
    data=titanic ,
    x="age" ,
)
```

```python
# So here now we will convert to standard normal variate
titanic["age"].mean() , titanic["age"].std()
```

```python
X = (titanic["age"] - titanic["age"].mean())/titanic["age"].std()
```

```python
sns.kdeplot(
    data=titanic,
    x=X ,
)
```
