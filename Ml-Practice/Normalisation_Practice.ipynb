```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(13, 5))


# note kde plot me x = df[] ese bhi le skte h pr nahi lna


sns.kdeplot(df['Class label'] , ax=ax1, color='blue', fill=0 )
sns.kdeplot(df['Malic acid'] , ax=ax1, color='red', fill=0 )


sns.kdeplot(df['Malic acid'] , ax=ax2, color='blue', fill=0 )
sns.kdeplot(df['Alcohol'] , ax=ax2, color='red', fill=0 )
sns.kdeplot(df['Class label'] , ax=ax2, color='green', fill=0 )
```

```python
color_dict = {1: 'red', 2: 'blue', 3: 'green'}  # Customize colors based on your class labels
# Create a scatter plot with different colors based on 'Class label'

sns.scatterplot(x='Alcohol', y='Malic acid', hue='Class label', palette=color_dict, data=df)

# Set title and labels
plt.title("Scatter Plot : Alcohol vs. Malic acid by Class")
plt.xlabel("Alcohol")
plt.ylabel("Malic acid")

# Display legend outside the plot

# plt.legend(title='Class', bbox_to_anchor=(1.50, 1), loc='upper left')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax
```

```python
df.columns
```

```python
df.dtypes
```

```python
minmax.fit(df[['Alcohol','Malic acid']])
```

```python
newdf = minmax.transform(df[['Alcohol','Malic acid']])
```

```python
dff = pd.DataFrame(newdf , columns = ['Alcohol' , 'Malic acid' ])
```

```python
dff
```

```python

clrs =  {1: 'red', 2: 'blue', 3: 'green'} 

sns.scatterplot(x = dff['Malic acid'] , y = dff['Alcohol'] , data = dff , color=clrs , hue = 'Alcohol' )
plt.title('Scatter Plot of Malic acid vs Alcohol')
plt.show()
```

```python

clrs =  {1: 'red', 2: 'blue', 3: 'green'} 

sns.scatterplot(x = df['Malic acid'] , y = df['Alcohol'] , data = df , color='red'  )
plt.title('Scatter Plot of Malic acid vs Alcohol')
plt.show()
```

# Kde For two data in one grapph 

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['Alcohol'], ax=ax1)
sns.kdeplot(df['Malic acid'], ax=ax1)



# after scaling
ax2.set_title('After Standard Scaling')

sns.kdeplot(dff['Alcohol'], ax=ax2)
sns.kdeplot(dff['Malic acid'], ax=ax2)


plt.show()
```

# Kde for only one in one graph

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['Alcohol'], ax=ax1)



# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(dff['Alcohol'], ax=ax2 , fill = 1 , color ='red')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 4))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['Malic acid'], ax=ax1 , fill = 1 , color = 'grey')


# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(dff['Malic acid'], ax=ax2 , fill = 1 , color ='red')
plt.show()
```
