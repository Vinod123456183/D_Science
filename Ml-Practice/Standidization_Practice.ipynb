```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

tips.head()
```

```python
tips.dtypes
```

```python

total_bill=list(tips['total_bill'])

sns.histplot(total_bill)
```

```python


import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)

```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std ## z_score
    normalized_data.append(z_score)
    
```

```python
sns.histplot(normalized_data)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot scatter plot for df on the first subplot (index 0)
sns.scatterplot(x='total_bill', y='tip', data=df, color='blue', ax=axs[0])
axs[0].set_title('Scatter Plot - DataFrame df')
axs[0].set_xlabel('Total Bill')
axs[0].set_ylabel('Tip')
axs[0].grid(True)



# Plot scatter plot for dff on the second subplot (index 1)
sns.scatterplot(x='total_bill', y='tips', data=dff, color='red', ax=axs[1])
axs[1].set_title('Scatter Plot - DataFrame dff')
axs[1].set_xlabel('Total Bill')
axs[1].set_ylabel('Tips')
axs[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(df['total_bill'], ax=ax1)
sns.kdeplot(df['tip'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(dff['total_bill'], ax=ax2)
sns.kdeplot(dff['tips'], ax=ax2)


plt.show()
```
