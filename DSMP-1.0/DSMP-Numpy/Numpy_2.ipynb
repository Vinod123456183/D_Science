```python
import numpy as np

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

actual,predicted
```

```python
def mse(actual,predicted):
  sum = 0
  for i in range(len(actual)):
      sum += (actual[i] - predicted[i])**2
  return sum/len(actual)
```

```python
ans = int(mse(actual,predicted))
```
