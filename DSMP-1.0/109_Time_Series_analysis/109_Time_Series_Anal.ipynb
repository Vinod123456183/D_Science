```python
import pandas as pd
from datetime import timedelta
import numpy as np
base_order_date = pd.to_datetime("2025-10-01")
order_dates = [base_order_date + timedelta(days=i) for i in range(10)]

delivery_dates = [date + timedelta(days=3) for date in order_dates]

# Create DataFrame
df = pd.DataFrame({
    'order_date': order_dates,
    'delivery_date': delivery_dates
})

df["delivery_time_period"] = df["delivery_date"]-df["order_date"]
```

```python

df["delivery_time_period"].mean()
```

```python
# Resampling = Changing the frequency of your time series observation

# Types

# Upsampling -> When u increase the frequency of the samples , such as from minutes to seconds
# downsampling -> when u decrease the freq of the samples such as from days to months
```

```python

# Every 7days ka mean le raha h bec resampling accepts aggregaete function later

google["Close"].plot()

# Increase value inside to view more clearly
google["Close"].resample("Y").mean().plot()
```

```python
# Its a rolling window . 

google["Close"].plot()
google["Close"].ewm(120).mean().plot()
google["Close"].rolling(120).mean().plot()
```
