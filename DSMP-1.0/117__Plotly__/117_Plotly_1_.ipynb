```python

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python

tips = px.data.tips()
iris = px.data.iris()
gap = px.data.gapminder()
```

```python
gap
```

```python

# Scatterplot using go

temp = gap[gap["year"] == 2007 ]
```

```python

trace1 = go.Scatter(x = temp["lifeExp"] , y = temp["gdpPercap"] , mode="markers")
trace2 = go.Scatter(x = [0,60,2] , y = [95,66,20555] , mode="lines")

data = [trace1 , trace2]

layout = go.Layout(title="Life Exp VS GDP per capita 2007 " , xaxis={"title":"Life Exp"} , yaxis={"title" : "GDP"})
fig = go.Figure(data,layout)
fig.show()
```

# TreeMap

```python
df = gap[gap["year"] == 2007]
px.treemap(
    df,
    path=[px.Constant("World") , "continent" , "country"] , values="pop" , color="lifeExp"
)
```

```python

# df = tips.pivot_table(index="day" , columns="sex" , values="total_bill" ,aggfunc="sum" )
df = gap.pivot_table(
    index="year",
    columns="continent",
    values = "lifeExp",
    aggfunc="mean"
)

px.imshow(
    df
)
```
