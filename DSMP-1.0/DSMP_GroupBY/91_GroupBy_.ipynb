```python
import pandas as pd
import numpy as np

df = pd.read_csv("IMDB top 1000.csv")
df.shape
```

```python
# Top 3 generes types movies that earn lots revenue

genre = df.groupby("Genre")
genre.sum()["Info"].sort_values(ascending=False).head(3)   #
genre["Info"].sum().sort_values(ascending=False).head(3)   # this is more better phle column then sum

```

```python
def rank_movie(grp):
    grp["Rank"] = grp["Rate"].rank(ascending=False)
    return grp
```

```python
df.groupby("Genre").apply(rank_movie)
```
