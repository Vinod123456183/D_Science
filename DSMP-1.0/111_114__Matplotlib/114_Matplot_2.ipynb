```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

iris = pd.read_csv("iris.data.csv", header=None)
iris.columns = ["SepalL", "SepalW", "PetalL", "PetalW", "Species"]
```

```python

plt.scatter(iris["SepalL"] , iris["PetalL"])
```

```python
iris["Species"] = iris["Species"].replace({"Iris-setosa":0 , "Iris-versicolor" : 1,"Iris-virginica":2})
```

```python

plt.scatter(iris["SepalL"] , iris["PetalL"] , c = iris["Species"] , cmap="jet")
plt.xlabel("Sepacl Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
# increase  Size

plt.figure(figsize=(13,8))
plt.scatter(iris["SepalL"] , iris["PetalL"] , c = iris["Species"] , cmap="jet")
plt.xlabel("Sepacl Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
bat = pd.read_csv("111/batter.csv")
```

```python

bat=bat.sample(25,random_state=5)

plt.scatter(bat["avg"] , bat["strike_rate"])
```

```python

# plt.xlabel("Average Runs")
# plt.ylabel("Strike Rate")

plt.figure(figsize=(15,10))
plt.scatter(bat["avg"] , bat["strike_rate"] , s=bat["runs"])

for i  in range(bat.shape[0]):
    plt.text(bat["avg"].values[i] , bat["strike_rate"].values[i] , bat["batter"].values[i])
    
```

```python
# Horizontal and vertical Lining


# plt.xlabel("Average Runs")
# plt.ylabel("Strike Rate")

plt.figure(figsize=(15,10))
plt.scatter(bat["avg"] , bat["strike_rate"] , s=bat["runs"])

plt.axhline(130,color="red")
plt.axvline(30,color="red")

for i  in range(bat.shape[0]):
    plt.text(bat["avg"].values[i] , bat["strike_rate"].values[i] , bat["batter"].values[i])

plt.show()
```

<h1>Subplots</h1>

```python

fig , ax = plt.subplots(figsize=(15,6))
ax.scatter(bat["avg"] , bat["strike_rate"],color="red")
ax.set_title("Something")
ax.set_xlabel("Something")
ax.set_ylabel("Somethingas")
fig.show()
```

```python

fig , ax = plt.subplots(nrows=2 , ncols=1 , figsize=(15,8))

ax[0].scatter(bat["avg"] , bat["strike_rate"] , color="red")
ax[1].scatter(bat["runs"] , bat["avg"] , color="blue")
ax[0].set_title("Avg VS Strike Rate")
ax[1].set_title("Runs VS Avg ")
```

```python

fig , ax = plt.subplots(nrows=2 , ncols=2 , figsize=(15,10))

ax[0][0].scatter(bat["avg"] , bat["strike_rate"] , color="red")
ax[0][1].scatter(bat["runs"] , bat["avg"] , color="blue")
ax[1][0].hist(bat["avg"] , color="green")
ax[1][1].hist(bat["runs"] , color="black")

ax[0][0].set_title("Avg VS Strike Rate")
ax[0][1].set_title("Runs VS Avg ")
ax[1][0].set_title("Avg ")
ax[1][1].set_title("Runs ")

```

<h1>3D Surface Plot</h1>

```python

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)
xx.shape , yy.shape

z = xx**2 + yy**2
z.shape


fig = plt.figure(figsize = (16,8))
ax = plt.subplot(projection="3d" )
p = ax.plot_surface(xx,yy,z , cmap="viridis")

fig.colorbar(p)
```

CounterPlot =>  Represnt 3d graph to 2D

```python

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx , yy = np.meshgrid(x,y)
xx.shape , yy.shape

z = xx**2 + yy**2
z.shape


fig = plt.figure(figsize = (13,6))
ax = plt.subplot( )
p = ax.contour(xx,yy,z , cmap="viridis")

fig.colorbar(p)
```
