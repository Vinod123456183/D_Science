<h1>BarChart</h1>


```python

childrens = [10,20,40,5,53]
colors = ["red" , "green" , "blue" , "lavendar" , "purple"]

plt.bar(colors , childrens , color="black")
```

```python
# if categories is increase then use horizontal
plt.barh(colors , childrens , color="black")
```

```python

df=pd.read_csv("111/batsman_season_record.csv")
# OverRiding
plt.bar(df["batsman"] , df["2015"])
plt.bar(df["batsman"] , df["2016"])
```

```python

plt.bar(df["batsman"] , df["2015"]  ,width=0.5)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)
```

```python
# Remove Existing labels with 0 1 2 3

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2 , color="blue")
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2 , color="green")
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color="red")

plt.xticks(np.arange(df.shape[0]), df["batsman"])

plt.show()
```

```python

childrens = [10,20,40,5,53]
colors = ["redredredredredred" , "greredredredre" , "bredredredlue" , "laredredredvendar" , "redredredredpurple"]

plt.bar(colors , childrens , color="black")
plt.xticks(rotation="vertical")
```

```python


# Stacked Barchart

```

```python

# Bottom me upar wala rakhte h and add krte rhna prta h

plt.bar(df["batsman"] , df["2017"] , label="2017")
plt.bar(df["batsman"] , df["2016"] , bottom=df["2017"] , label="2016" )
plt.bar(df["batsman"] , df["2015"] , bottom=(df["2016"] + df["2017"]) , label="2015" )
plt.legend()
plt.show()
```
