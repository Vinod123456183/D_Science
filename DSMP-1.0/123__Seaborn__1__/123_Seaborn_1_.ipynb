```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

tips = sns.load_dataset("tips")
```

```python

# Here we have not given label automatically 
    # in matplotlib we have to give labesl

# Axis level function
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    style="time",
    hue="sex",
    size="size"
)
```

```python
# Figure level Function
    # Generally are square in shape
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="scatter",
    style="time",
    hue="sex",
    size="size"
)
```

```python

gap=px.data.gapminder()
india = gap[gap["country"] == "India"]

sns.relplot(
    data=india,
    x="year",
    y="pop",
    kind="line",   
)
```

```python
ipc = gap[gap["country"].isin(["India" , "Pakistan" , "China"])]
ipc
```

```python

# Figure level me legend  chiz bahar hoti h

sns.relplot(
    data=ipc,
    x="year",
    y="pop",
    kind="line",
    hue="country"
)
```

```python
ieg = gap[gap["country"].isin(["India" , "Brazil" , "Germany"])]
ieg


# Figure level me legend  chiz bahar hoti h

sns.relplot(
    data=ieg,
    x="year",
    y="pop",
    kind="line",
    hue="country",
    style="continent",
    size="continent"
)
```

```python

# Facet Plot
```

```python
sns.relplot(
    data=tips,
    x = "total_bill",
    y="tip",
    kind="scatter",
    col="sex",
    row="smoker",
    hue="sex",
)
```

```python

sns.relplot(
    data=gap,
    x = "lifeExp",
    y="gdpPercap",
    kind="scatter",
    col="year",
    col_wrap=3
)
```

# 2 Distributed Plots

```python
sns.displot(
    data=tips,
    x="tip",
    hue="sex",
    kind="hist",
    bins=10
)
```

```python
# We can also plot in categoral value in sns 


sns.displot(
    data=tips,
    x="day",
    hue="sex",
    col="sex"
)
```

```python
sns.displot(
    kind="kde",
    x="total_bill",
    data=tips,
    hue="sex",
    fill=True
)
```

```python
sns.displot(
    kind="kde",
    x="total_bill",
    data=tips,
    hue="sex",
    fill=True
)
sns.rugplot(
    x="total_bill",
    data=tips,
)
```

```python

# Bivariet histogram

sns.displot(
    data=tips,
    x="total_bill",
    y="tip",
    kind ="hist"
)
```
