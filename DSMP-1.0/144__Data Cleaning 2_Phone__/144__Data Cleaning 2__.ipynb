```python

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("updated_smart_phones.csv")
df.head()
```

```python
brand_names = df["model"].str.split(" ").str.get(0)
```

```python
df.insert(1,"Brand Names" , brand_names)
df.head(4)
```

```python
df["Brand Names"] = df["Brand Names"].str.lower()
df.head()
```

```python

has_5g = df["sim"].str.contains("5G")
has_nfc = df["sim"].str.contains("NFC")
has_ir_blaster = df["sim"].str.contains("IR Blaster")
```

| Method            | Use When...                                       | Example Match             |
| ----------------- | ------------------------------------------------- | ------------------------- |
| `.str.contains()` | You want to check if a **substring** exists       | "5G" in "Dual SIM, 5G"    |
| `.isin()`         | You want to check if the **entire value** matches | exact match: "5G" == "5G" |


```python

df.insert(6 , "Has 5G" , has_5g)
df.insert(7 , "Has NFC" , has_nfc)
df.insert(8 , "Has IR Blaster" , has_ir_blaster)
```

```python
df.head(3)
```

```python

processor_name = df["processor"].str.split(',').str.get(0)
no_of_cores = df["processor"].str.split(',').str.get(1)
processor_speed = df["processor"].str.split(',').str.get(2)
df.head(2)
```

```python
df.insert(10,"Processor Name",processor_name)
df.insert(11,"No Of Cores",no_of_cores)
df.insert(12,"Processor Speed",processor_speed)
```

```python

df.head(1)
```

```python


df["Processor Name"] = df["Processor Name"].str.strip()
```

```python
temp = df[df["Processor Name"].str.contains("Core")][["Processor Name"	,"No Of Cores"	,"Processor Speed"]].shift(1,axis = 1)
temp
```

```python
# Editing multi row and col so we give index and row col 

df.loc[temp.index,["Processor Name" , "No Of Cores" , "Processor Speed"]]  = temp.values
df["Processor Name"].value_counts()
```

```python
processor_brand = df["Processor Name"].str.split(" ").str.get(0).str.lower()
df.insert(11,"Processor Brand",processor_brand)
df.head(2)
```

```python

df["No Of Cores"] = df["No Of Cores"].str.strip()
df["No Of Cores"].value_counts()
```

```python

df["No Of Cores"] = df["No Of Cores"].str.replace("Octa Core Processor","Octa Core").str.replace("Hexa Core Processor","Hexa Core")
# i was storing value counts
df["No Of Cores"].value_counts()
```

```python
df["Processor Speed"] = df["Processor Speed"].str.strip().str.split(" ").str.get(0).str.replace("\u2009" , " ").str.split(" ").str.get(0).astype(float)
```

```python
df["ram"].str.strip().str.split(",").str.get(0)[0]
```

```python
ram_capicity_in_GB = df["ram"].str.strip().str.split(",").str.get(0).str.replace("\u2009" , " ").str.split(" ").str.get(0)

df.columns

df.insert(16 , "Ram Capicity (GB)" , ram_capicity_in_GB)
df.head(2)
```
