# Imp Graph For ML

```python

# 3D Surface Plot is not in plotly express
# Plotly Graph objects

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)


xx,yy = np.meshgrid(x,y)

z=xx**2 + yy**2
# z=np.sin(xx) + np.cos(yy)

trace = go.Surface(
    x = x,
    y = y,
    z = z
)

layout = go.Layout(title = "3D Surface Plot")
data=[trace]

fig=go.Figure(
    data,
    layout
)

fig.show()
```

```python

# 3D Surface Plot is not in plotly express
# Plotly Graph objects

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)


xx,yy = np.meshgrid(x,y)

z=xx**2 + yy**2
# z=np.sin(xx) + np.cos(yy)

trace = go.Contour(
    x = x,
    y = y,
    z = z
)

layout = go.Layout(title = "3D Surface Plot")
data=[trace]

fig=go.Figure(
    data,
    layout
)

fig.show()
```

# Matplot Subplot VS Plotly Subpllot 

```python
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2,
    cols=2
)
```

```python
fig.add_trace(
    go.Scatter(x = [1,9,5] , y = [5,9,3]),
    row=1,
    col=1
)

fig.add_trace(
    go.Histogram(x = [1,9,55,9,3.55,5,34,354,354,35,43,54]),
    row=1,
    col=2,
)


fig.add_trace(
    go.Scatter(x = [21,39,55] , y = [55,39,53]),
    row=2,
    col=1
)

fig.add_trace(
    go.Histogram(x = [1,9,55,9,3,56,5,34,35,4,3,54,35,43,54]),
    row=2,
    col=2,
)


fig.update_layout(title = "Subplot Demo")
fig.show()
```
