```python

import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python

# Stirplot => ek numerical , ek categorical 

sns.stripplot(
     tips,
    x="day",
    y="total_bill",
    hue="sex"
)
```

```python

# Using Fig level function
sns.catplot(
    tips,
    x="day",
    y="tip",
    kind="strip"
)
```

```python

# SwarmPlot
 # Better representation bec internal algo runs  behind it 
    # Better for small datasets

# Using Fig level function
sns.catplot(
    tips,
    x="day",
    y="tip",
    kind="swarm"
)
```
