```python
Data1 = pd.DataFrame({
    'Key1' : [1,2,3,4],
    'Key2' : [7,8,9,11],
    'Key3' : [9,7,8,9]
})
```

```python
Data2 = pd.DataFrame({
    'Key1' : [1,2,3,4],
    'Key2' : [7,8,9,111],
    'Key5' : [101,202,302,44]
})
```

```python
Data1
```

```python
Data2
```

```python
pd.merge(Data1,Data2)
```

```python
pd.merge(Data1,Data2 , how = 'left')

# Sb Kuch uthao Data1 se fir Common element k stah merge kro wrna nan bhar do

```

```python
pd.merge(Data1,Data2 , how = 'right')

# Sb Kuch uthao Data2 se fir Common element k stah merge kro wrna nan bhar do

```

```python
pd.merge(Data1,Data2 , how = 'outer')

# dono jagah se 
```

```python
pd.merge(Data1,Data2 , how = 'cross')

# do Cartesian Product
```
