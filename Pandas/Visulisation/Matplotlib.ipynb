# plots between categrotical and numerical

```python
x=['a' , 'q' , 'r' , 'g' , 'p']

y= np.random.rand(5)
```

```python
plt.bar(x,y)
plt.xlabel("x - Axis")
plt.ylabel("y - Axis")
plt.title("Bar-plot")
plt.show()
```

```python
plt.barh(x,y)
plt.xlabel("x - Axis")
plt.ylabel("y - Axis")
plt.title("Bar-plot")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(x , y ,'--g' )
plt.xlabel("x - Axis")
plt.ylabel("y - Axis")
plt.title("Rln bt x and y plot")

plt.show()
```
