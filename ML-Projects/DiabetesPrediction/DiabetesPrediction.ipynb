```python
# Sepreating Data And label

Y = df['Outcome']
X = df.drop(columns = 'Outcome' , axis = 1)
```

Standardise  === In Standardise We can use fit_transform ek sath 

```python

scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardise_data = scaler.transform(X)
```

```python
X = standardise_data
Y = df['Outcome']
```

```python
X
```

```python
# Train Test
```

```python

X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , stratify=Y ,  random_state=2)

X.shape , X_train.shape , X_test.shape
```

```python
from sklearn import svm
classifier = svm.SVC(kernel = 'linear')
```

```python
classifier.fit(X_train , Y_train)
```

```python

X_train_prediction = classifier.predict(X_train)

training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Training Data Acuracy : ',training_data_accuracy)
```

```python

X_test_prediction = classifier.predict(X_test)

test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Training Data Acuracy : ',test_data_accuracy)
```
