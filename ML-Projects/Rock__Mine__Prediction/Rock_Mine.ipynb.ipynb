```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Give Header if none value given
df = pd.read_csv('Sonar_Mine.csv' , header=None)
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
print(X )
```

```python

X_train , X_test , Y_train , Y_test = train_test_split(X , Y  , test_size= .2 , random_state=4)
```

```python
X_train.shape , X_test.shape
```

```python

print(X_train)
print(Y_train)
```

```python
X_train.shape , Y_train.shape , Y_train.shape , X.shape , Y_train.shape , Y_test.shape
```

```python

lr = LogisticRegression()
```

```python
#training the Logistic Regression model with training data

lr.fit(X_train,Y_train)
```

```python
# LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
#                    intercept_scaling=1, l1_ratio=None, max_iter=100,
#                    multi_class='auto', n_jobs=None, penalty='l2',
#                    random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
#                    warm_start=False)
```

```python
# note while calculating acc score \ cross value put , Y_test
X_train_predict = lr.predict(X_train)
Accuraccy_Score_Training = accuracy_score(X_train_predict , Y_train)
print("Accuraccy Score In Training Data is :",Accuraccy_Score_Training * 100)
```

```python
X_test_predict = lr.predict(X_test)
Acc_Score_Test = accuracy_score(X_test_predict , Y_test)
print("Accuracy Score on Test Data :",Acc_Score_Test*100)
```
